# Exploratory Data Analysis (EDA) - Heart Disease Dataset
This notebook fulfills all EDA requirements for the MLOps Assignment 01.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load data
df = pd.read_csv('../data/heart.csv')
display(df.head())
```

## 1. Missing Value Analysis
Check for any missing values in the dataset.

```
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

print("Missing values count per column:")
print(df.isnull().sum())
```

## 2. Class Distribution Plots
Check if the target variable (heart disease presence) is balanced.

```
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='target', palette='Set2')
plt.title('Target Class Distribution')
plt.show()
```

## 3. Histograms
Distribution of numerical features.

```
df.hist(figsize=(15, 12), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of All Features', y=1.02)
plt.tight_layout()
plt.show()
```

## 4. Correlation Heatmap
Analyze correlations between features and the target variable.

```
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

## 5. Feature Relationship Analysis
Boxplots to see how specific features relate to the target class (e.g. max heart rate and age vs target).

```
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.boxplot(ax=axes[0], data=df, x='target', y='thalach', palette='Set1')
axes[0].set_title('Max Heart Rate (thalach) vs Target')

sns.boxplot(ax=axes[1], data=df, x='target', y='age', palette='Set1')
axes[1].set_title('Age vs Target')
plt.show()
```
